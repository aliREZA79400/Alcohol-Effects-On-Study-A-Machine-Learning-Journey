## Common Attributes (Both Datasets)

| Column        | Description & Values |
|---------------|-----------------------|
| `school`      | Student's school (binary: `GP` - Gabriel Pereira, `MS` - Mousinho da Silveira) |
| `sex`         | Student's gender (binary: `F` - Female, `M` - Male) |
| `age`         | Student's age (numeric: 15-22) |
| `address`     | Home address type (binary: `U` - Urban, `R` - Rural) |
| `famsize`     | Family size (binary: `LE3` ≤ 3 members, `GT3` > 3 members) |
| `Pstatus`     | Parent's cohabitation status (binary: `T` - Together, `A` - Apart) |
| `Medu`        | Mother's education (numeric: 0=None, 1=Primary, 2=5th-9th Grade, 3=Secondary, 4=Higher) |
| `Fedu`        | Father's education (numeric: same scale as Medu) |
| `Mjob`        | Mother's occupation (nominal: `teacher`, `health`, `services`, `at_home`, `other`) |
| `Fjob`        | Father's occupation (nominal: same categories as Mjob) |
| `reason`      | Reason for school choice (nominal: `home`, `reputation`, `course`, `other`) |
| `guardian`    | Student's guardian (nominal: `mother`, `father`, `other`) |
| `traveltime`  | Home to school travel time (numeric: 1=<15min, 2=15-30min, 3=30min-1hr, 4=>1hr) |
| `studytime`   | Weekly study time (numeric: 1=<2hrs, 2=2-5hrs, 3=5-10hrs, 4=>10hrs) |
| `failures`    | Number of past class failures (numeric: 0-4, where 4 means ≥3 failures) |
| `schoolsup`   | Extra educational support (binary: `yes`, `no`) |
| `famsup`      | Family educational support (binary: `yes`, `no`) |
| `paid`        | Extra paid classes **within subject course** (binary: `yes`, `no`) |
| `activities`  | Extra-curricular activities (binary: `yes`, `no`) |
| `nursery`     | Attended nursery school (binary: `yes`, `no`) |
| `higher`      | Wants higher education (binary: `yes`, `no`) |
| `internet`    | Home internet access (binary: `yes`, `no`) |
| `romantic`    | Romantic relationship (binary: `yes`, `no`) |
| `famrel`      | Family relationship quality (numeric: 1=Very bad - 5=Excellent) |
| `freetime`    | Free time after school (numeric: 1=Very low - 5=Very high) |
| `goout`       | Going out with friends (numeric: 1=Very low - 5=Very high) |
| `Dalc`        | Workday alcohol consumption (numeric: 1=Very low - 5=Very high) |
| `Walc`        | Weekend alcohol consumption (numeric: same scale as Dalc) |
| `health`      | Current health status (numeric: 1=Very bad - 5=Very good) |
| `absences`    | Number of school absences (numeric: 0-93) |

## Course-Specific Grades

### Maths.csv
| Grade | Description |
|-------|-------------|
| `G1`  | First period grade in Mathematics (numeric: 0-20) |
| `G2`  | Second period grade in Mathematics (numeric: 0-20) |
| `G3`  | **Final grade in Mathematics** (numeric: 0-20, target variable) |

### Portuguese.csv
| Grade | Description |
|-------|-------------|
| `G1`  | First period grade in Portuguese (numeric: 0-20) |
| `G2`  | Second period grade in Portuguese (numeric: 0-20) |
| `G3`  | **Final grade in Portuguese** (numeric: 0-20, target variable) |

**Note:**  
- Target variable for prediction: `G3` (final grade)  
- Original data source: Cortez & Silva, 2008 (UCI Machine Learning Repository)

**Note:**

*This data approach student achievement in secondary education of two Portuguese schools. The data attributes include student grades, demographic, social and school related features) and it was collected by using school reports and questionnaires. Two datasets are provided regarding the performance in two distinct subjects: Mathematics (mat) and Portuguese language (por). In [Cortez and Silva, 2008], the two datasets were modeled under binary/five-level classification and regression tasks. Important note: the target attribute G3 has a strong correlation with attributes G2 and G1. This occurs because G3 is the final year grade (issued at the 3rd period), while G1 and G2 correspond to the 1st and 2nd period grades. It is more difficult to predict G3 without G2 and G1, but such prediction is much more useful (see paper source for more details).*

```python
from scipy.stats import chi2_contingency
import numpy as np

def perform_chi_square_test(df, var1, var2, alpha=0.05):
    # Create contingency table
    contingency = pd.crosstab(df[var1], df[var2])
    
    # Perform chi-square test
    chi2, p_value, dof, expected = chi2_contingency(contingency)
    
    # Print results
    print(f"\nChi-Square Test: {var1} vs {var2}")
    print(f"Chi-Square Statistic: {chi2:.4f}")
    print(f"p-value: {p_value:.4f}")
    print(f"Degrees of Freedom: {dof}")
    print("Significant relationship" if p_value < alpha else "No significant relationship")
    print("-" * 50)
    
    return p_value

# Get all categorical columns
cat_cols = stu_cat.columns.tolist()

# Store results
results = []

# Perform chi-square test for each pair of categorical variables
for i in range(len(cat_cols)):
    for j in range(i+1, len(cat_cols)):
        var1, var2 = cat_cols[i], cat_cols[j]
        p_value = perform_chi_square_test(stu_cat, var1, var2)
        results.append({
            'Variable 1': var1,
            'Variable 2': var2,
            'p_value': p_value
        })

# Create DataFrame of results and sort by p-value
results_df = pd.DataFrame(results)
results_df = results_df.sort_values('p_value')

# Display top 10 most significant relationships
print("\nTop 10 Most Significant Relationships:")
print(results_df.head(10))

# Visualize one of the significant relationships
significant_pair = results_df.iloc[0]
var1, var2 = significant_pair['Variable 1'], significant_pair['Variable 2']

# plt.figure(figsize=(12, 6))
# contingency = pd.crosstab(stu_cat[var1], stu_cat[var2])
# sns.heatmap(contingency, annot=True, fmt='d', cmap='YlOrRd')
# plt.title(f'Contingency Table: {var1} vs {var2}')
# plt.show()
```

```python
stu_labels = stu["G3"].copy()
stu_data = stu.copy().drop(columns=["G1","G2","G3","reason","activities","guardian","famsize","famsup","traveltime"])
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()

mjb = ohe.fit_transform(stu_cat["Mjob"].to_frame()).toarray()
fjb = ohe.fit_transform(stu_cat["Fjob"].to_frame()).toarray()
```

```python
def or_operation_binary(arr1, arr2):
    """
    Performs an element-wise OR operation on two NumPy arrays containing 0 and 1,
    returning a new array with 0 and 1 values. Handles various input types.

    Args:
        arr1: The first NumPy array.
        arr2: The second NumPy array.

    Returns:
        A new NumPy array with the result of the OR operation (0s and 1s).
        Raises ValueError if input arrays are not the same shape or contain values that cannot be interpreted as 0 or 1.
    """

    if arr1.shape != arr2.shape:
        raise ValueError("Input arrays must have the same shape.")

    # Convert to integers, handling potential errors:
    try:
        arr1 = arr1.astype(int)
        arr2 = arr2.astype(int)
    except ValueError:
        raise ValueError("Input arrays must contain values that can be interpreted as 0 or 1.")

    # Check for valid values *after* type conversion:
    if not (np.all(np.isin(arr1, [0, 1])) and np.all(np.isin(arr2, [0, 1]))):
        raise ValueError("Input arrays must contain only 0 and 1 values (after type conversion).")


    result = np.bitwise_or(arr1, arr2).astype(int)
    return result

or_operation_binary(mjb,fjb)
```

```python
from sklearn.compose import make_column_selector ,ColumnTransformer
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from scipy.sparse import csr_matrix
from sklearn.preprocessing import FunctionTransformer ,StandardScaler 

def column_or(X):
    # print("column_or ...")
    #TODO temperory solution ,solve after
    if X.shape[1] == 10 :
        return csr_matrix(or_operation_binary(X[:,0:5].toarray()
                                          ,X[:,5:10].toarray()))
    else :
        return csr_matrix(or_operation_binary(X[:,0:2].toarray()
                                          ,X[:,2:4].toarray()))

def or_name(function_transformer,feature_names_in):
    return ["or"]

def or_pipeline():
    # print("or_pipeline ...")
    return make_pipeline(
        SimpleImputer(strategy="most_frequent"),
        OneHotEncoder(handle_unknown="ignore"),
        FunctionTransformer(column_or,feature_names_out=or_name)
    )

cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore")
)
# Numeric columns
def column_sum(X):
    # print("column_sum ...")
    # print(X.shape)
    return np.sum(X,axis=1).reshape(-1, 1)

def sum_name(function_transformer,feature_names_in):
    return ["sum"]

def sum_pipeline(): 
    # print("sum_pipeline ...")
    return make_pipeline(
        SimpleImputer(strategy="median"),
        FunctionTransformer(column_sum,feature_names_out=sum_name),
        StandardScaler()
    )
log_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    FunctionTransformer(np.log1p , feature_names_out='one-to-one'),
    StandardScaler()
)

num_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler()
)
###

preprocessing = ColumnTransformer([
    ("MFjob", or_pipeline(), ["Mjob","Fjob"]),
    ("SH", or_pipeline(), ["schoolsup","higher"]), 
    ("cat", cat_pipe, make_column_selector(dtype_include='object')),
    ("DW", sum_pipeline(), ["Dalc","Walc"]),
    ("MFedu", sum_pipeline(), ["Medu","Fedu"]),
    ("gDWff", sum_pipeline(), ["Dalc","Walc","goout","freetime","failures"]),
    ("log", log_pipeline, ["Dalc","failures","absences","age"]),
    ("num", num_pipeline, ["Medu","Fedu","studytime","famrel","freetime",
                          "goout","Walc","health"])
])

```

## Quick look and finding best model

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

linear_reg = make_pipeline(preprocessing,
                          LinearRegression())
linear_reg.fit(stu_data, stu_labels)

linear_err_train = mean_squared_error(stu_labels ,
                             linear_reg.predict(stu_data))


print(f"Train Error {np.sqrt(linear_err_train)}")
```

```python
from sklearn.model_selection import cross_val_score

linear_cross_val = -cross_val_score(linear_reg , stu_data,stu_labels,
                              scoring="neg_root_mean_squared_error",
                              cv=10)
pd.Series(linear_cross_val).describe()
```

### DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = make_pipeline(preprocessing,DecisionTreeRegressor(random_state=42))
tree_reg.fit(stu_data, stu_labels)
tree_err_train = mean_squared_error(stu_labels ,
                             tree_reg.predict(stu_data))



print(f"Train Error {tree_err_train}")
```

```python

tree_cross_val = -cross_val_score(tree_reg , stu_data,stu_labels,
                              scoring="neg_root_mean_squared_error",
                              cv=10)
pd.Series(tree_cross_val).describe()
```

### RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = make_pipeline(preprocessing,
                           RandomForestRegressor(random_state=42))

forest_rmses = -cross_val_score(forest_reg, stu_data, stu_labels,
scoring="neg_root_mean_squared_error", cv=10)

pd.Series(forest_rmses).describe()
```

## Model selection

```python

def evaluate_model_tradeoff(model, model_name, X, y, cv=10):
    scores = cross_val_score(model, X, y, 
                           scoring='neg_root_mean_squared_error',
                           cv=cv)
    return {
        'Model': model_name,
        'Mean_RMSE': np.mean(-scores),
        'Std_RMSE': np.std(-scores),
        'CV_Scores': -scores
    }

# Evaluate models
results = [
    evaluate_model_tradeoff(linear_reg, "Linear", stu_data, stu_labels),
    evaluate_model_tradeoff(tree_reg, "Decision Tree", stu_data, stu_labels),
    evaluate_model_tradeoff(forest_reg, "Random Forest", stu_data, stu_labels)
]

# Create comparison plot
plt.figure(figsize=(10, 6))
for res in results:
    plt.scatter(res['Mean_RMSE'], res['Std_RMSE'], s=200, alpha=0.7,
               label=res['Model'])
    plt.errorbar(res['Mean_RMSE'], res['Std_RMSE'],
                 xerr=res['Std_RMSE']/2, yerr=res['Std_RMSE']/2,
                 alpha=0.3)

plt.xlabel('Mean RMSE (Lower Better)', fontsize=12)
plt.ylabel('Std RMSE (Lower Better)', fontsize=12)
plt.title('Model Trade-off: Mean vs Stability', fontsize=14)
plt.legend()
plt.grid(True)

# Add optimal quadrant guidance
plt.axvline(x=np.mean([res['Mean_RMSE'] for res in results]), 
           color='gray', linestyle='--', alpha=0.5)
plt.axhline(y=np.mean([res['Std_RMSE'] for res in results]), 
           color='gray', linestyle='--', alpha=0.5)
plt.text(plt.xlim()[1]*0.6, plt.ylim()[0]*1.1, 
        'Ideal Zone →', color='green', fontsize=12)

plt.show()
```
